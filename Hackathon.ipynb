```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, confusion_matrix, accuracy_score
```

```python
x = df.drop(columns=['salary'])
```

```python
y = df['salary']
```

```python
Encoder_x = LabelEncoder()
for col in x.columns:
    x[col]=Encoder_x.fit_transform(x[col])
Encoder_y = LabelEncoder()
```

```python
x.head()
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
y_train.head()
```

```python
# KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train,y_train)
knn_predict = knn.predict(x_test)
accuracy_score(knn_predict,y_test)
```

```python
#Decision_tree
dst_tree = DecisionTreeClassifier()
dst_tree.fit(x_train,y_train)
dst_predict = dst_tree.predict(x_test)
accuracy_score(dst_predict,y_test)
```

```python
#Random_forest
ran_forest = RandomForestClassifier()
ran_forest.fit(x_train,y_train)
ran_forest_predict = ran_forest.predict(x_test)
accuracy_score(ran_forest_predict,y_test)
```

```python
#Gaussian

gau = GaussianNB()
gau.fit(x_train,y_train)
gau_predict = gau.predict(x_test)
accuracy_score(gau_predict,y_test)
```
